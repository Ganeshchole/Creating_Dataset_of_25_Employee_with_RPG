```python
class employee:
    def __init__(self):
        self.count = int(input('enter how many employees data you have to fill : '))
        self.id = []
        self.nme = []
        self.gen = []
        self.age = []
        self.city = []
        self.salary = []
        self.total_exp = []
        self.company_exp = []

    def emp_Data(self):
        x=0
        while x < self.count:                   
            self.id.append(int(input("id : ")))               
            self.nme.append(input("name : "))
            self.gen.append(input("gender : "))
            self.age.append(int(input("age : ")))                
            self.city.append(input("city :"))
            self.total_exp.append(int(input("total_exp : ")))
            self.company_exp.append(int(input("emp_of_comp : ")))              
            self.salary.append(int(input("salary : ")))
            print("- "*20)
            x += 1
            
        db = {"id": self.id, 
               "name": self.nme, 
               "gender": self.gen, 
               "age": self.age, 
               "city": self.city, 
               "salary": self.salary, 
               "total_exp": self.total_exp, 
               "emp_of_comp": self.company_exp
              }
        
        import pandas as pd
        df = pd.DataFrame(db)
        df.to_csv("emp.csv")
        print(df)
        return df
```

```python
ob = employee()
```

```python
data=ob.emp_Data()
```
